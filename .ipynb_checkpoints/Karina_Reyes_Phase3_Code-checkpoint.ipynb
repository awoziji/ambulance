```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
```

```python
#Answer to 1

# Reading the data into a DataFrame
ambdata = pd.read_csv('annual_ambulance_data.csv',encoding = 'ISO-8859-1')
ambdata = ambdata.set_index('LGA')

profiles = pd.read_csv('lga_profile.csv',encoding = 'ISO-8859-1')
profiles = profiles.set_index('LGA')

lgadata = pd.concat([ambdata, profiles], axis=1)
lgadata
```

```python
#Answer to 2
lgadata['Pop_Den'] = lgadata['Population']/lgadata['Area_of_LGA_sq_km']
lgadata['Hospital_Den'] = lgadata['Number_of_hospitals_and_health_']/lgadata['Area_of_LGA_sq_km']
lgadata.sort_values('demand')
lgadata['Percent_Under15mins'] = lgadata['Percent_Under15mins']

highdemand = lgadata[lgadata['demand']>6.5]
lowrt = highdemand[highdemand['Percent_Under15mins']<0.6]
ratio = 100*len(lowrt)/len(highdemand)
print(ratio)
```

```python
#Answer to 3
cities=lgadata.loc[lgadata["ARIA_Remoteness_category"]=='Major Cities of Australia']
inner_region=lgadata.loc[lgadata["ARIA_Remoteness_category"]=='Inner Regional Australia']
outer_region=lgadata.loc[lgadata["ARIA_Remoteness_category"]=='Outer Regional Australia']

a='Distance_to_Melbourne_in_km'
b='AVG_RT_Seconds'

plt.scatter(cities.loc[:,a],cities.loc[:,b],color='green', label='Major Cities')
plt.scatter(inner_region.loc[:,a],inner_region.loc[:,b],color='red', label = 'Inner Regional VIC')
plt.scatter(outer_region.loc[:,a],outer_region.loc[:,b],color='blue', label = 'Outer Regional VIC')
#plt.xlim(0.5,8100)
#plt.ylim(0.5,8.0)
plt.legend() #Show legend mapping colors to days of the week
plt.ylabel("Ambulance Response Time (Seconds)")
plt.xlabel("Area of LGA (Sq. Km)")
plt.title("Scatter Plot of LGA Area vs Code 1 Response Times")
plt.show()

```

```python
lgadata['demand'] = lgadata['Total_Calls']/(lgadata['Population']/100)
lgadata
```

```python
#Answer to 4
hospitals = lgadata["demand"].tolist()
lganames = lgadata.index

plt.bar(arange(79), hospitals, width = 0.5)
plt.ylim(0,10)
plt.xlim(-1, 79)
plt.xticks(arange(79), lganames, rotation = 90)
plt.ylabel("Calls per 100 people")
plt.xlabel("LGA")
plt.title("Ambulance Calls per 100 People")
plt.figure(figsize=(1,1))


# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print("Current size:", fig_size)
 
# Set figure width to 12 and height to 9
fig_size[0] = 8
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size

plt.show()
```
